```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 

def extract_color(event,x,y,flags,param) :
    if event==cv2.EVENT_LBUTTONDOWN : 
        lower_bound=hsv_im[y,x].astype(np.uint8)
        higher_bound=lower_bound+np.array([20,20,20]).astype(np.uint8)
        print(higher_bound)
        print(lower_bound)
        mask=cv2.inRange(hsv_im,lower_bound,higher_bound)
        result=cv2.bitwise_and(im,im,mask=mask)
        cv2.imshow("result",result)
        





im=cv2.imread("bunny.jpeg")
hsv_im=cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
cv2.namedWindow("image") 
cv2.setMouseCallback("image",extract_color)

while True : 

    cv2.imshow("image",im)

    key=cv2.waitKey(1)
    if key==27 : 
        break 

cv2.destroyAllWindows() 
```
